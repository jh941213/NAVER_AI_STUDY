- Ï§ëÎ≥µÎêòÎäî Î¨ºÌíàÏùÄ Ìï©Ï≥êÏÑú ÌëúÏãúÌïòÏÑ∏Ïöî.
- Í∞Å ÎîïÏÖîÎÑàÎ¶¨ Îç∞Ïù¥ÌÑ∞Ïùò Îç∞Ïù¥ÌÑ∞Ïùò ÌÇ§Í∞íÏùÑ Ïù¥Ïö©Ìï¥ Ï§ëÎ≥µÏùÑ ÌôïÏù∏Ìï¥Î≥¥ÏÑ∏Ïöî.

```python
dict_first = {'ÏÇ¨Í≥º': 30, 'Î∞∞': 15, "Í∞ê": 10, "Ìè¨ÎèÑ":10}
dict_second = {'ÏÇ¨Í≥º': 5, 'Í∞ê': 25, "Î∞∞": 15, "Í∑§":25}

def merge_dict(dict_first, dict_second):
    # ÏßëÌï© ÏûêÎ£åÌòïÏúºÎ°ú Ï§ëÎ≥µ Í≥ºÏùº Ïù¥Î¶Ñ Ï†úÍ±∞ ( )
    fruit_list = set(dict_first.keys()) | set(dict_second.keys())
    # Í∞íÏùÑ 0ÏúºÎ°ú Í∞ñÎäî ÎîïÏÖîÎÑàÎ¶¨ ÏÉùÏÑ±
    dict_result = {fruit:0 for fruit in fruit_list}
    for fruit,price in dict_result.items():
        # Ï≤´ Î≤àÏß∏ ÎîïÏÖîÎÑàÎ¶¨Ïóê Í≥ºÏùºÏù¥ Ï°¥Ïû¨ÌïòÎ©¥ Í∞íÏùÑ ÎçîÌï®
        if fruit in dict_first:
            dict_result[fruit] += dict_first[fruit]
        # Îëê Î≤àÏß∏ ÎîïÏÖîÎÑàÎ¶¨Ïóê Í≥ºÏùºÏù¥ Ï°¥Ïû¨ÌïòÎ©¥ Í∞íÏùÑ ÎçîÌï®
        if fruit in dict_second:
            dict_result[fruit] += dict_second[fruit]

    # Î¨∏Ï†úÏóêÏÑú ÏöîÍµ¨ÌïòÎäî ÌòïÌÉúÎ°ú ÎßûÏ∂∞Ï£ºÍ∏∞ ÏúÑÌï¥ key Ï†ïÎ†¨ Î∞è dictionary ÏûêÎ£åÌòïÏúºÎ°ú Î∞òÌôò
    return dict(sorted(dict_result.items()))


merge_dict(dict_first, dict_second)
```

## Q5. Îã®Ïñ¥Îì§ÏùÑ ÏûÖÎ†•Î∞õÏïòÎäîÎç∞, ÏûêÍæ∏Îßå Ïà´ÏûêÎì§Ïù¥ ÏÑûÏó¨Îì§Ïñ¥Í∞ÄÎäî Î¨∏Ï†úÍ∞Ä ÏûàÏäµÎãà Îã§. Ïù¥Î•º Ï≤òÎ¶¨ÌïòÍ∏∞ ÏúÑÌï¥ÏÑú Ìï®ÏàòÏóê stringÏùÑ Ï†ÑÎã¨ Î∞õÏùÄ Îí§, string ÏïàÏóêÏÑúÏùò Ïà´ ÏûêÎ•º Ï†úÍ±∞Ìïú ÌõÑ stringÎßå ÎÇ®ÏùÄ Î¶¨Ïä§Ìä∏Î•º Ï∂úÎ†•ÌïòÏÑ∏Ïöî. üòé

- string Ïó∞ÏÇ∞ÏùÑ Ïù¥Ïö©Ìï¥ÏÑú Î¨∏ÏûêÏó¥ÏùÑ ÏûêÎ•¥Îäî Ïó∞ÏÇ∞ÏùÑ ÏÇ¨Ïö©Ìï¥Î≥¥ÏÑ∏Ïöî.

```python
# Ï£ºÏñ¥ÏßÑ ÏûÖÎ†• string
inputs = "cat32dog16cow5"

def find_string(inputs):
    # Ïà´ÏûêÍ∞Ä ÏïÑÎãàÎ©¥ listÎ°ú Î≥ÄÌôò
    input_list = [word for word in inputs if not word.isdigit()]
    # 3 Ïπ∏Ïî© Ïù∏Îç±Ïä§Î•º ÎùÑÏõåÍ∞ÄÎ©∞ string Î≥ëÌï©ÌïòÍ≥† listÎ°ú Î≥ÄÌôòÌïòÏó¨ Î∞òÌôò
    input_list = [input_list[idx] + input_list[idx+1] + input_list[idx+2] for idx in range(0,len(input_list),3)]
    # Îòê Îã§Î•∏ ÏÑ†ÌÉùÏßÄ
    # Ï†ïÍ∑úÌëúÌòÑÏãù ÏÇ¨Ïö©
    # ASCII ÏΩîÎìúÍ∞í ÏÇ¨Ïö©
    return input_list

string_list = find_string(inputs)
print(string_list)
```
